# **Show the three color channels (R, G, B).**

• Separate the image into its three color channels (Red, Green, and Blue) and display them
individually, labeling each channel as R, G, and B.{Using NumPy.}

```python
# Step 1: Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Step 2: Open the image and convert it to a NumPy array
image = Image.open("test.png")  # Replace with your image file
image_array = np.array(image)    # Convert image to NumPy array

# Step 3: Extract individual R, G, B channels
red_channel = image_array.copy()   # Copy original image array
green_channel = image_array.copy()
blue_channel = image_array.copy()

# Step 4: Keep only one color channel at a time
red_channel[:, :, 1:] = 0    # Set Green & Blue to 0, keep Red
green_channel[:, :, [0, 2]] = 0  # Set Red & Blue to 0, keep Green
blue_channel[:, :, :2] = 0   # Set Red & Green to 0, keep Blue

# Step 5: Display all three color channels
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].imshow(red_channel)   # Show Red channel
axes[0].set_title("Red Channel (R)")
axes[0].axis("off")

axes[1].imshow(green_channel)  # Show Green channel
axes[1].set_title("Green Channel (G)")
axes[1].axis("off")

axes[2].imshow(blue_channel)  # Show Blue channel
axes[2].set_title("Blue Channel (B)")
axes[2].axis("off")

plt.show()
```

# **Extract and display the middle section of the image (150 pixels).**

• Extract a 150 pixel section from the center of the image using NumPy array slicing.

• Display this cropped image using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Step 1: Open the image and convert it to a NumPy array
image = Image.open("cameraman.png")  # Replace with your image file
image_array = np.array(image)  # Convert image to NumPy array

# Step 2: Get image dimensions (height and width)
height, width = image_array.shape[:2]  # Get height and width; ignore channels if present

# Step 3: Calculate center region for 150-pixel crop
center_x, center_y = width // 2, height // 2  # Get center coordinates
half_size = 75  # Half of 150 pixels

# Step 4: Extract the 150x150 center region
center_crop = image_array[center_y - half_size : center_y + half_size,
                          center_x - half_size : center_x + half_size]

# Step 5: Display the cropped middle section
plt.imshow(center_crop, cmap='gray' if len(image_array.shape) == 2 else None)
plt.axis("off")  # Hide axes
plt.show()
```

# **Apply a simple threshold to the image (e.g., set all pixel values below 100 to 0).**

• Apply a threshold to the grayscale image: set all pixel values below 100 to 0, and all values
above 100 to 255 (creating a binary image).

• Display the resulting binary image.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Step 1: Open the image and convert it to a grayscale image
image = Image.open("cameraman.png").convert("L")  # Convert to grayscale
image_array = np.array(image)  # Convert image to NumPy array

# Step 2: Apply thresholding
threshold = 100
binary_image = np.where(image_array < threshold, 0, 255).astype(np.uint8)  # Create binary image

# Step 3: Display the resulting binary image
plt.imshow(binary_image, cmap='gray')  # Display as grayscale
plt.axis("off")  # Hide axes
plt.show()
```

# **Convert the grayscale image to an RGB image.**

• Convert the grayscale image into an RGB image where the grayscale values are replicated
across all three channels (R, G, and B).

• Display the converted RGB image using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Step 1: Open the grayscale image
image = Image.open("cameraman.png").convert("L")  # Ensure the image is in grayscale
image_array = np.array(image)  # Convert image to NumPy array

# Step 2: Convert grayscale to RGB by stacking the grayscale values
rgb_image_array = np.stack((image_array,) * 3, axis=-1)  # Replicate across 3 channels

# Step 3: Convert back to an image
rgb_image = Image.fromarray(rgb_image_array.astype(np.uint8))

# Step 4: Display the converted RGB image
plt.imshow(rgb_image)
plt.axis("off")  # Hide axes
plt.show()

```

# **Eigen Decomposition and Identifying Principal Components:**

• Compute Eigen Values and Eigen Vectors.

• Sort the eigenvalues in descending order and choose the top k eigenvectors corresponding to
the highest eigenvalues.

• Identify the Principal Components with the help of cumulative Sum plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Step 1: Load and prepare data
image = Image.open("lena_gray.gif").convert("L")  # Replace with your image file
image_array = np.array(image)  # Convert image to NumPy array
flattened_image = image_array.flatten()

# Center the data
mean = np.mean(flattened_image)
centered_data = flattened_image - mean
centered_data_2d = centered_data.reshape(image_array.shape[0], image_array.shape[1])
covariance_matrix = np.cov(centered_data_2d, rowvar=False)

# Step 2: Compute Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Step 3: Sort the eigenvalues and eigenvectors
sorted_indices = np.argsort(eigenvalues)[::-1]  # Indices for sorting in descending order
sorted_eigenvalues = eigenvalues[sorted_indices]  # Sorted eigenvalues
sorted_eigenvectors = eigenvectors[:, sorted_indices]  # Corresponding eigenvectors

# Step 4: Choose top k eigenvectors
k = 10  # Number of principal components to keep
top_eigenvectors = sorted_eigenvectors[:, :k]

# Step 5: Identify Principal Components with cumulative sum plot
explained_variance = sorted_eigenvalues / np.sum(sorted_eigenvalues)  # Explained variance
cumulative_variance = np.cumsum(explained_variance)  # Cumulative sum of explained variance

# Step 6: Plot cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.title('Cumulative Explained Variance by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.xlim(1, len(cumulative_variance))
plt.ylim(0, 1)
plt.axhline(y=0.95, color='r', linestyle='--')  # Line at 95% threshold
plt.text(len(cumulative_variance), 0.95, '95% Threshold', color='red', verticalalignment='bottom')
plt.show()

# Print results
print("Top k Eigenvalues:\n", sorted_eigenvalues[:k])
print("Top k Eigenvectors (first 5 values):\n", top_eigenvectors[:5])
```

# **Reconstruction and Experiment:**

**• Reconstruction:** Transform the original data by multiplying it with the selected eigenvec-
tors(PCs) to obtain a lower-dimensional representation.

**• Experiments:** Pick Four different combination of principal components with various ex-
plained variance value and compare the result.

**• Display the Results and Evaluate.**

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Step 1: Load and prepare data
image = Image.open("lena_gray.gif").convert("L")  # Replace with your image file
image_array = np.array(image)
flattened_image = image_array.flatten()

# Center the data
mean = np.mean(flattened_image)
centered_data = flattened_image - mean
centered_data_2d = centered_data.reshape(image_array.shape[0], image_array.shape[1])
covariance_matrix = np.cov(centered_data_2d, rowvar=False)

# Step 2: Compute Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Ensure we are working with real values
eigenvalues = np.real(eigenvalues)
eigenvectors = np.real(eigenvectors)

# Step 3: Sort Eigenvalues and Eigenvectors
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

# Step 4: Function to reconstruct the image
def reconstruct_image(num_components):
    # Select the top k eigenvectors
    top_eigenvectors = sorted_eigenvectors[:, :num_components]

    # Transform the data to lower dimensions
    transformed_data = np.dot(centered_data_2d, top_eigenvectors)

    # Reconstruct the image
    reconstructed_data = np.dot(transformed_data, top_eigenvectors.T) + mean  # Add mean back
    return np.clip(reconstructed_data, 0, 255).astype(np.uint8)  # Ensure pixel values are in valid range

# Step 5: Experiment with different numbers of principal components
components_to_test = [5, 10, 30, 70]  # Different combinations of principal components
reconstructed_images = []

for k in components_to_test:
    reconstructed_image = reconstruct_image(k)
    reconstructed_images.append(reconstructed_image)

# Step 6: Display the results
plt.figure(figsize=(12, 8))

# Original image
plt.subplot(2, 3, 1)
plt.imshow(image_array, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Reconstructed images
for i, k in enumerate(components_to_test):
    plt.subplot(2, 3, i + 2)
    plt.imshow(reconstructed_images[i], cmap='gray')
    plt.title(f'Reconstructed Image with {k} PCs')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
